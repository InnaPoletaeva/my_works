```python
# Подключаем библиотеки
import requests
from bs4 import BeautifulSoup
```

```python
# Указываем адрес страницы, с которой будем собирать данные и отправляем запрос
book = 'https://mybook.ru/catalog/malaya-forma/books/free/?cursor=cj0xJnA9MC4wNzYyMjgxNDM2MzYwOTAx'
result = requests.get(book)
```

```python
# Проверка кода
result.status_code
```

```python
# Просматриваем текст ответа
result.text
```

```python
# Применяем конструктур BeautifulSoup для разбора ответа
soup = BeautifulSoup(result.text, 'lxml')
```

```python
# Смотрим, что получилось
print(soup.prettify())
```

```python
# Ищем нужный тег с нужными параметром и значением
storys = soup.find_all('div', attrs = {'class' : ['e4xwgl-0 iJwsmp']})
len(storys)
```

```python
# Смотрим один фрагмент кода - книгу из списка с индексом 0
storys[0]
```

```python
# Пытаемся найти необходимые сведения для выбранного фрагмента кода страницы
# Воспользуемся методом get_text(strip = True), указанный параметр
# возращает текст без спецсимволов

# Найдем автора
storys[0].find('div', attrs = {'class':'dey4wx-1 jVKkXg'}).get_text(strip = True)
```

```python
# Найдем название книги
storys[0].find('p', attrs = {'class':'lnjchu-1 hhskLb'}).get_text(strip = True)
```

```python
# Найдем рейтинг книги
storys[0].find('span', attrs = {'class':'sc-1vvnv6o-3 hTOIIb'}).get_text(strip = True)
```

```python
# Найдем платно или бесплатно читать книгу
storys[0].find('p', attrs = {'lnjchu-1 kgTBVx sc-3u14ee-1 OSKeK'}).get_text(strip = True)
```

```python
# Найдем информацию о книге
storys[0].find('p', attrs = {'lnjchu-1 dPgoNf'}).get_text(strip = True)
```

```python
# Склеиваем, чтобы получить правильную ссылку, тег 'a' и параметр 'href'
'https://mybook.ru/' + storys[0].find('a').get('href')
```

```python
# Создаем спискок для хранения результатов, где каждый элемент содержит одну книгу
list_storys = list()

# В словать будем помещать информацию об одной книге, а дальше этот словарь будем писать в список
dict_story = dict.fromkeys(['author', 'title', 'rating', 'paid or free', 'about the book', 'url'], 0)
```

```python
# С помощью цикла обрабатываем найденные записи
for item in storys:
    #информацию о каждой книге пишем в словарь
    dict_story['author'] = item.find('div', attrs = {'class':'dey4wx-1 jVKkXg'}).get_text(strip = True)
    dict_story['title'] = item.find('p', attrs = {'class':'lnjchu-1 hhskLb'}).get_text(strip = True)
    dict_story['rating'] = item.find('p', attrs = {'class':'lnjchu-1 kgTBVx'}).get_text(strip = True)
    dict_story['paid or free'] = item.find('p', attrs = {'lnjchu-1 kgTBVx sc-3u14ee-1 OSKeK'}).get_text(strip = True)
    dict_story['about the book'] = item.find('p', attrs = {'lnjchu-1 dPgoNf'}).get_text(strip = True)
    dict_story['url'] = 'https://mybook.ru/' + storys[0].find('a').get('href')
    #словрь пишем в список
    list_storys.append(dict_story.copy())
```

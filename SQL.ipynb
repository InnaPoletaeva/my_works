```python
# Подключаем библиотеку
import pandas as pd
import sqlite3
```

```python
# Подключаемся к базе и создаем курсор

con = sqlite3.connect("/content/book_simple.db.db")
cur = con.cursor()
```

```python
# Сделаем запрос на информацию по таблицам в БД

request_info = """
SELECT name FROM sqlite_master
WHERE type='table'
ORDER BY name;
"""
```

```python
# Выполним запрос, видим, что БД состоит из 5 таблиц, как на картинке выше

res = cur.execute(request_info)
res.fetchall()
```

```python
# Посмотрим одну из таблиц, например book

show_book = """
SELECT *
FROM book;
"""
pd.read_sql(show_book, con)
```

```python
# 1. Выведите авторов, имеющихся в базе.

aut = """
SELECT name
FROM author;
"""
pd.read_sql(aut, con)
```

```python
# 2. Найдите всех авторов с инициалами "С. А.".
# Автор в базе определяется так: И. О. Фамилия

aut_1 = """
SELECT name
FROM author
WHERE name LIKE "С. А. %";
"""
pd.read_sql(aut_1, con)
```

```python
# 3. Выведите список книг, выпущенных в издательстве "Мир".

book_mir = """
SELECT book.title,
       publisher.name
FROM book, publisher
WHERE (publisher.id = book.publisher_id) AND (publisher.name = 'Мир')
"""
pd.read_sql(book_mir, con)
```

```python
# 4. Выведите количество книг, принадлежащих каждой из категорий.
# Результат отсортировать по возрастанию.

book_col_cat = """
SELECT category.title,
       COUNT (book.id) AS 'Количество книг'
FROM category
INNER JOIN book
ON category.id = book.category_id
GROUP BY category.title
ORDER BY 2 ASC;
"""
pd.read_sql(book_col_cat, con)
```

```python
# 5. Выведите количество книг, выпущенных в каждом издательстве.
# В итоговом списке оставьте только те издательства, которые выпустили 3 и более книг.
# Результат отсортировать по убыванию.
```

```python
book_col_publ = """
SELECT publisher.name,
       COUNT (publisher.name) AS 'Количество_книг'
FROM publisher
INNER JOIN book
ON publisher.id = book.publisher_id
GROUP BY publisher.name
HAVING Количество_книг >= 3
ORDER BY 2 DESC;
"""
pd.read_sql(book_col_publ, con)
```

```python
# 6. Выведите список книг, автором которых являлся <<С. А. Айвазян>>.

book_caa = """
SELECT author.name,
       book.title
FROM book
INNER JOIN authorbook
ON book.id = authorbook.book_id
INNER JOIN author
ON author.id = authorbook.author_id
WHERE author.name = 'С. А. Айвазян'
"""
pd.read_sql(book_caa, con)
```

```python
# 7. В каких издательствах и сколько выпускались книги за авторством <<С. А. Айвазян>>.
# Результат отсортировать по убыванию.

publ_col = """
SELECT publisher.name AS 'Издательство',
       COUNT (publisher.name) AS 'Количество_книг',
       author.name AS 'Автор'
FROM publisher
INNER JOIN book
ON publisher.id = book.publisher_id
INNER JOIN authorbook
ON book.id = authorbook.book_id
INNER JOIN author
ON author.id = authorbook.author_id
WHERE author.name = 'С. А. Айвазян'
GROUP BY publisher.name
ORDER BY 2 DESC;
"""
pd.read_sql(publ_col, con)
```
